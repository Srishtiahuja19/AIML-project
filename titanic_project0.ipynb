```python
def fill_age(row):
  Pclass=row[0]
  age=row[1]
  if pd.isnull(age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 28
    elif Pclass==3:
      return 22
  else:
    return age

```

```python
titanic_data[['Pclass','Age']].apply(add_10,axis=1)
```

```python
titanic_data['Age']=titanic_data[['Pclass','Age']].apply(fill_age,axis=1)
```

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.reset_index(drop=True,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder()
```

```python
one_hot.fit(titanic_data[['Embarked']])
```

```python
temp_arr=one_hot.transform(titanic_data[['Embarked']]).toarray()
```

```python
temp_arr
```

```python
temp_data=pd.DataFrame(temp_arr,columns=['C','Q','S'],dtype='int')
```

```python
temp_data
```

```python
titanic_data=pd.concat([titanic_data,temp_data],axis=1)
```

```python
titanic_data.drop('Embarked',axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
X=titanic_data.drop('Survived',axis=1)
Y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split  #complusary

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
predictions=logistic.predict(X_test)
```

```python
Y_test[9]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(predictions,Y_test))
```

```python
print(classification_report(predictions,Y_test))
```

```python
sns.heatmap(confusion_matrix(predictions,Y_test,annot=True))
```
